## TF-IDF, basically
----

###TF = term-frequency, or just word counts on a per-document basis

- boolean: tf(w, d) = 1 if w in d, else 0
- raw: tf(w,d) = number of times w appears in d
- log: tf(w,d) = $\lg(raw + 1)$, where +1 is smoothing constant
- and other, fancier schemes

I started with raw counts. More on that later

###IDF = inverse-document-frequency, or how often does this appear across the corpus

- raw: $$idf(w) = \frac{N}{C(w \in D) + 1}$$
- log: $idf(w) = \lg(raw)$

Mentions seemed to only use log version, which is what I went with

###TF-IDF = the combination of TF and IDF...
Very simply: $tfidf(w, d) = tf(w,d) * idf(w)$

#Then what?
***

Queries - "show me documents most similar to $\{some \;text\}$"
Measure similarity with cosine
$$\cos(\theta) = \frac{A \cdot B}{||A||\;||B||} = \frac{\sum A_iB_i}{\sqrt{\sum A_i^2} * \sqrt{\sum B_i^2}}$$
where e.g. A is the TF-IDF of a search query, and B is a given TF-IDF vector (column in our TF-IDF table)

Query 1: "I am all that"

Query 2: "Ends become beginnings"

**TF-IDF**

| |Query1 |Query2|
|--|--|--|
|all| .4 | 0|
|that|.4 | 0|
|ends| 0| 1|

**Cosine Similarity**

| |Doc 1|Doc 2|Doc 3|Doc 4|
|--|--|--|--|
|Query1|$$\frac{.4*.4 + .4*.4}{\sqrt{.4^2+.4^2}*\sqrt{.4^2+.4^2}} = 1$$ |$$\frac{.4*.4}{\sqrt{.4^2+.4^2}*\sqrt{.4^2}} = 0.71$$ |$$\frac{.4*.4}{\sqrt{.4^2+.4^2}*\sqrt{.4^2}} = 0.71$$|$$\frac{0}{\sqrt{.4^2+.4^2}*\sqrt{0}} =0 $$|
|Query2|$$\frac{1*1}{\sqrt{1^2}*\sqrt{1^2}} = 1$$ |$$\frac{0}{\sqrt{1^2}*\sqrt{0}} = 0$$|$$\frac{0}{\sqrt{1^2}*\sqrt{0}} = 0$$|$$\frac{0}{\sqrt{1^2}*\sqrt{0}} = 0$$|

Tells us:

- Query 1 is most similar to Document 1, somewhat similar to Documents 2 and 3.
- Query 2 is similar to Document 1, and not to any of the others

## Eventual Hypothesis
---

If we can suggest appropriate tags to users *as they are writing the question*, perhaps we can get more tags, and more accurate tags, applied to a question. 

Before I can see if people are more likely to apply tags if given a tag-suggestion list, I need to first derive the tag-suggestion list.

## This Hypothesis
---
Using the textual information provided by a user, provide a list of tag suggestions that are likely to be the most relevant to the question the user is proposing.

Aspects and approach:

- Just use title and text, not user metadata
- Perform some preprocesing of data
- Relatively quick runtime, so suggestions can be provided during submission


## First thoughts:
Let's use the TF-IDF I spent so much time explaining

The dataset I chose consisted of 5000 questions, which contain **title**, **text**, **tags** for each question.

Perhaps the title of the question holds different data than the text-body? Title words might be more direct, more concise.

##Approach:
Perform TF-IDF on all the titles and separately, all the texts. 

We would then have an $NxM_{title}$ matrix, with $N$ titles and $M_{title}$ unique words across all titles, 

and an $NxM_{text}$ matrix, with $N$ texts and $M_{text}$ unique words across all texts.

- **Interjection**: Normalize term-frequency by document length! Shortest title: 2 words, longest: 29 words. Shortest text: 9 words, longest: 1948 words

$$tf(w, d) = \frac{raw\_count(w)}{len(d)}$$

If we stopped here, a Query would return the **title** (or the whole question, if desired). But what we eventually want is a tag-suggestion list.


##Tag-specific TF-IDF
---

Imagine we only had 1 tag. Our goal is to figure out how likely it is that this tag should be applied to the query (which is a question, ie title-text pairing). And what do we have at our disposal? In this contrived example, we have TF-IDF matrices for all of our training questions, and (somewhere) the knowledge of whether or not our 1 tag had been applied to that training question.

If we were to toss this query at these matrices, it would compute a cosine similarity for each and every training question, and then we could start to look at, say, whether the most similar question had that tag applied or not. However, that may not quite tell us what we want - what if all of the next closest questions have an opposite classification? 

Instead, let's look at it a bit differently: we want to know whether this Query is more or less similar to questions that typically have this tag, to help decide whether or not this tag should be applied.

So, we can combine all the columns of our TF-IDF matrix where the tag has been applied, to get overall TF-IDF "trends" about that particular tag. 

- **Interjection**: this combination should be an average, not a sum, of the cell values. Otherwise, you start overweighting frequently applied tags and all your experiments go haywire for a day or two...

Now, we've reduced each (title or text) entire TF-IDF matrix to a single column, where the column represents the average of all documents which had the tag applied. 

Calculations just got much simpler: $tfidf(query, tag-tfidf)$ will yield a single cosine-similarity score (noting, again, that you have two of these, one for the *tag-title-tfidf*, the other for the *tag-text-tfidf*)

The natural extension, then, is to build this tag-specific tf-idf for each and every tag. Is this really that big of a win, over just doing TF-IDF on each of the questions? For the dataset I used, there were 5000 questions, and only 100 tags. Maintaining information only for the tag-tfidf instead of for each question is then likely to yield about an order of magnitude improvement in memory usage, and in time spent computing cosine similarities. For the largest StackOverflow sub-site, there are over 7 million questions, and over 30,000 tags.

##What Have We Done?
---

We turned this:

| | $question_1$ | $question_2$| ... | $question_N$|
|--|-|-|-|-|
|$tag_1$|x| |...| x|
|$tag_2$| |x|...| |
|...|...|...|$\ddots$|...|
|$tag_K$|x | x|...| |
|$word_1$|.4|.1|...|0|
|$word_2$|.1|.1|...|.999|
|...|...|...|$\ddots$|...|
|$word_M$|.2|.8|...|.1|

...into...

| |$tag_1$|$tag_2$|...|$tag_K$|
|-|-|-|-|-|
|$word_1$|.1|.2|...|.8|
|$word_2$|.4|.4|...|.2|
|...|...|...|$\ddots$|...|
|$word_M$|.6|.3|...|.99|

which is much smaller - $M \times K$ vs. $N\times (M+K)$

and more relevantly, gets us very close to what we were originally intending: now, a query will return a vector of cosine similarities based on how relevant each *tag* is to the query. If we sort these results, we can grab the highest scoring results according to some selection function, and that would be our tag suggestion list.

More formally, and to be somewhat lame and claim I'm doing fancy Machine Learning...

Define a distance function on two TF-IDF vectors (e.g. $tfidf(query, tag_i)$ ) to be:
$$dist(v_1, v_2) = 1 - tfidf(v_1, v_2)$$
We can then (easily and trivially) convert our vector of cosine similarities to a vector of distances.

Then define some neighborhood function such as with *k-nearest-neighbors* that takes in the vector of distances, and returns the $k$ closest. Voila, I used k-nearest neighbors! Trivially, but hey!

# Experiments
---

We are now ready to actually do some experiments on this data:

The trivial example functions we could consider and compare against are:

- random: choose tags out of the tag list randomly
- popularity: choose the most frequent tags for every result

I would hope and expect both of these to do poorly, otherwise there'd be no need for my application.

What kind of selection functions can we look at? Recall that for a query, we have a vector of tags, where each element contains a title-similarity and a text-similarity for the given tag.

- title-only: $f(query, tag) = tfidf(query_{title}, tag_{title})$
- text-only: $f(query, tag) = tfidf(query_{text}, tag_{text})$
- linear combo: $f(query, tag) = tfidf(query_{title}, tag_{title}) + tfidf(query_{text}, tag_{text})$
- nonlinear: $f(query, tag) = tfidf(query_{title}, tag_{title}) * tfidf(query_{text}, tag_{text})$

This should give us a decent base. 

What sort of neighborhood parameter should we use? 

- 5 tags: that's the maximum upper limit on tag selection
- 10 tags: they're just suggestions and there's no penalty for ignored suggestions (to an extent)
- threshold on the distance: (future work, when I have a better algorithm)

What sort of evaluation metric?
$$sensitivity = \frac{TP}{TP+FN}$$
What's a $True Positive$? When a tag that has been applied to a question is contained within the tag suggestion list.

Whats a $False Negative$? When a tag that has been applied to a question is **not** contained within the tag suggestion list.

As a result, a single query can result in multiple $TP$s and $FN$s.

Why not $specificity$? We know we are vastly overshooting - on average, a post has 2 tags, and we are suggesting either 5 or 10. specificity rates will be abysmal (future work, when I have a better algorithm...)

## So, *Less* Data?
---

We are stuck in this interesting position of, at least for the data sizes chosen, less data is better. But... we've... got more data. What's an algorithm to do? 

Simplest answer: split up the training data into different (non-overlapping) partitions, train each TF-IDF for each tag on each partition, and then aggregate the results and see if it's any better. Rationally should this work? As rationally as ensembles work, because that's at heart what this amounts to.

- **Interjection**: Normally, an ensemble method such as bagging would train the first classifier (or in this case TF-IDF) with some subset of data, then the next classifier would get some of the misclassifications and some other training data. However, in a multi-output problem such as this, it isn't straight-forward how to deal with this. Is it a misclassification if it misses one of multiple tags? Or only if it misses *all* tags? If you reuse a datapoint, you need to keep all of its tags associated with it, not just the ones misclassified, since otherwise you'll be training on incorrect data... To avoid thinking about all this, I just strictly partitioned the data. A clever approach though could yield some nice results.

How many different partitions should we create? Let's try a few different values. Arbitrarily, I investigated 3, 5, and 20. This number (hereafter referred to as $J$) is open to metaparameter analysis, but I didn't have the time or compute power to do a thorough investigation.

Note that the experiments performed above are essentially $J=1$, which is to say "an ensemble of 1 TF-IDF classifier". 

```python
width=.22
ds, js, ns, sizes = [[0, 4, 8, 12], [33, 37, 41, 45]], [1, 3, 5, 20], [5, 10], [1000, 4000]
cs = ["b", "g", "r", "pink"]
fig, axes = plt.subplots(ncols=2, figsize=figsize(16, 5))
fig.subplots_adjust(wspace=.5)
for aidx, ax in enumerate(axes):
    title = "$D={}$, $n={}$".format(sizes[aidx], ns[aidx])
    df = [G.data[i][4:8] for i in ds[aidx]]
    ind = np.arange(len(df[0]))
    for idx in range(len(df)):
        ax.bar(ind+(idx*width), df[idx], width=width, color=cs[idx], label="$J="+str(js[idx])+"$")
    ax.set_xticks(ind+width)
    ax.set_xticklabels(labels[:4], fontsize=16, rotation=45)
    ax.set_xlabel("function", fontsize=20)
    ax.set_ylabel("sensitivity", fontsize=20)
    ax.set_ylim(.25, .8)
    ax.set_title(title, fontsize=30)
    ax.legend(loc=8)
None # output killer
```
